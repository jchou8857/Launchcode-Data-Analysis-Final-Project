## Visualize

Create any visualizations for your EDA here. Make note in the form of code comments of what your thought process is for your visualizations.

```python
# Another look at how types of awards are distributed. Used percentage of total awards given to better display this.

plt.figure(figsize=(10,5))
sns.histplot(x=df2['Award'], stat='percent').set(title='Award Type as Percentage of Total Awards')
```

```python
# Visualize how the locations of the restaurants are distributed.

sns.countplot(data=df2, y='Location', order=df2['Location'].value_counts().iloc[:25].index).set(title='Top 25 Locations')
```

```python
# Visualize how the cuisines of the restaurants are distributed.

sns.countplot(data=df2, y='Cuisine', order=df2['Cuisine'].value_counts().iloc[:25].index).set(title='Top 25 Cuisines')
```

```python
x = 'Award'
y = 'Cuisine'

# Compute the counts of observations
df_counts = df2.groupby([x, y]).size().reset_index()
df_counts.columns.values[df_counts.columns == 0] = 'count'

# Compute a size variable for the markers so that they have a good size regardless
# of the total count and the number of unique values in each categorical variable
scale = 2*df_counts['count'].size
size = df_counts['count']/df_counts['count'].sum()*scale

# Create matplotlib scatter plot with additional formatting
fig, ax = plt.subplots(figsize=(10,300))
ax.scatter(x, y, size, data=df_counts, zorder=2)
ax.grid(color='grey', linestyle='-', alpha=0.4, zorder=1)
ax.tick_params(length=0)
ax.set_frame_on(False)
ax.margins(x=0.3, y=0.001)
```
