```python
# Importing Pandas to clean data.
# Importing SQLite3 for easier querying of data.
# Importing Matplotlib, Matplotlib.pyplot, Seaborn, and Numpy for visualiztion.

import pandas as pd
import sqlite3 as sl
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Reading data.
df1 = pd.read_csv('Michelin Guide Data_cleaned.csv')
df1
```

```python
# Separating out multiple cuisine values for a count.
df1[['Cuisine1','Cuisine2']] = df1['Cuisine'].str.split(', ',expand=True)
df1
```

```python
df1_1 = df1[['Cuisine1']]
df1_1.rename(columns={'Cuisine1': 'CuisineSeparated'},inplace=True)
df1_2 = df1[['Cuisine2']]
df1_2.rename(columns={'Cuisine2': 'CuisineSeparated'},inplace=True)
df_cuisines = pd.concat([df1_1, df1_2], ignore_index=True)
df_cuisines_count = df_cuisines.value_counts()
df_cuisines_count
```

```python
# Reading info into a separate csv file:
df1.to_csv('Michelin Guide Data_Separated_Cuisines.csv', index=False)
df_cuisines_count.to_csv('Michelin Guide Data_Cuisine_Count.csv')
```
